```julia
A(Z) = (1.06Z^(-0.38) + 0.18)*1E-3
B(Z) = 0.22Z^(-0.055) + 0.78
C(Z) = (1.1Z^(-0.29) + 0.21)*1E-3
R_e_ext(E_k; Z=32) = A(Z)*E_k*(1-B(Z)/(1+C(Z)*E_k))
plot(0.01:10, [E -> R_e_ext(E*1E3, Z=32)*10/ge_density, E -> R_e_ext(E*1E3, Z=18)*10/lar_density],
     title="Electron extrapolated range", lw=2,
     labels=["germanium" "LAr"], legend=:top,
     xaxis=("Energy (MeV)", :log),
     yaxis=("Range (mm)", :log10),
     minorgrid=true)
```

```julia
savefig("electron-ext-range.pdf")
```

```julia
R_alpha_air(E_k) = 0.31E_k^(3/2)
R_alpha(E_k; A=76, rho=ge_density) = (1E4*3.2E-4sqrt(A)/rho)*R_alpha_air(E_k)
plot(1:10, [E -> R_alpha(E, rho=ge_density), E -> R_alpha(E, rho=lar_density), E -> R_alpha_air(E)/air_density],
     title="alpha particles range", lw=2,
     labels=["germanium" "LAr" "air"], legend=:best,
     xaxis=("Energy (MeV)"),
     yaxis=("Range (microns)", :log10),
     minorgrid=true)
```
